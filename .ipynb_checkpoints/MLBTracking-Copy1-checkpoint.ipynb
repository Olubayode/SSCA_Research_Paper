## Quality Reports of Data

```python
# using pandas
def quality_report(df):
    """
    Description: Displays quality of data in terms of missing values, 
    unique numbers, datatypes etc.
    
    Arguments: Dataframe
    """
    dtypes = df.dtypes
    nuniq = df.T.apply(lambda x: x.nunique(), axis=1)
    total = df.isnull().sum().sort_values(ascending = False)
    percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
    quality_df  = pd.concat([total, percent, nuniq, dtypes], axis=1, keys=['Total NaN', 'Percent of NaN','Nunique', 'Dtype'])
    display(quality_df)
```

```python
import polars as pl

def quality_report_pl(df: pl.DataFrame) -> pl.DataFrame:
    n_rows = df.height

    # total nulls per column
    null_counts = (
        df.select([pl.col(c).is_null().sum().alias(c) for c in df.columns])
          .melt(variable_name="column", value_name="total_nulls")
          .with_columns( (pl.col("total_nulls") * 100.0 / n_rows).alias("percent_nulls") )
    )

    # n_unique per column
    nunique = (
        df.select([pl.col(c).n_unique().alias(c) for c in df.columns])
          .melt(variable_name="column", value_name="n_unique")
    )

    # dtypes
    dtypes = pl.DataFrame({
        "column": list(df.schema.keys()),
        "dtype":  [str(t) for t in df.schema.values()],
    })

    q = (
        null_counts
        .join(nunique, on="column", how="left")
        .join(dtypes,  on="column", how="left")
        .with_columns([
            pl.col("total_nulls").cast(pl.Int64),
            pl.col("percent_nulls").cast(pl.Float64),
            pl.col("n_unique").cast(pl.Int64),
        ])
        .sort(["percent_nulls", "total_nulls"], descending=True)
    )
    return q

# usage
qr = quality_report_pl(df)
qr.head()
print(qr.head)
```

```python
import polars as pl

def quality_report_with_counts(df, cols=None):
    if cols is None:
        cols = df.columns
    n_rows = df.height

    # nulls per column
    nulls = (
        df.select([pl.col(c).is_null().sum().alias(c) for c in cols])
          .melt(variable_name="column", value_name="null_count")
    )

    # nans per column (only for float dtypes)
    exprs = []
    for c in cols:
        if df.schema[c] in pl.FLOAT_DTYPES:          # robust float check
            exprs.append(pl.col(c).is_nan().sum().alias(c))
        else:
            exprs.append(pl.lit(0).alias(c))
    nans = df.select(exprs).melt(variable_name="column", value_name="nan_count")

    # n_unique per column
    nunique = (
        df.select([pl.col(c).n_unique().alias(c) for c in cols])
          .melt(variable_name="column", value_name="n_unique")
    )

    # dtypes
    dtypes = pl.DataFrame({"column": cols, "dtype": [str(df.schema[c]) for c in cols]})

    out = (
        nulls.join(nans, on="column")
             .join(nunique, on="column")
             .join(dtypes, on="column")
             .with_columns([
                 (pl.lit(n_rows) - pl.col("null_count")).alias("non_nulls"),
                 (pl.col("null_count") * 100.0 / n_rows).alias("percent_nulls"),
             ])
             .select(["column","dtype","non_nulls","null_count","percent_nulls","nan_count","n_unique"])
             .sort(["percent_nulls","null_count"], descending=True)
    )
    return out

# usage
qr = quality_report_with_counts(df_clean, ["swing_length", "bat_speed"])
print(qr)
```

## Stability check (ARI) across threshold

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

# helper: cluster at a threshold
def cluster_at_T(df_clean: pl.DataFrame, T: int):
    agg = (df_clean.group_by("batter")
                  .agg([
                      pl.count().alias("n_swings"),
                      pl.mean("bat_speed").alias("avg_bat_speed"),
                      pl.mean("swing_length").alias("avg_swing_length"),
                  ])
                  .filter(pl.col("n_swings") >= T))
    X = agg.select(["avg_bat_speed","avg_swing_length"]).to_pandas().values
    bat = agg.select("batter").to_pandas().values.ravel()
    km = KMeans(n_clusters=3, n_init=20, random_state=42)
    labels = km.fit_predict(X)
    return pl.DataFrame({"batter": bat, f"label_{T}": labels})

labs_50  = cluster_at_T(df_clean, 50).to_pandas()
labs_100 = cluster_at_T(df_clean, 100).to_pandas()
labs_150 = cluster_at_T(df_clean, 150).to_pandas()

# compare on intersecting hitters
def ari(a, b, k1, k2):
    m = a.merge(b, on="batter", how="inner")
    return adjusted_rand_score(m[k1], m[k2]), len(m)

ari_50_100, n_50_100 = ari(labs_50, labs_100, "label_50", "label_100")
ari_100_150, n_100_150 = ari(labs_100, labs_150, "label_100", "label_150")
print(f"50 vs 100: ARI={ari_50_100:.3f} over {n_50_100} common hitters")
print(f"100 vs 150: ARI={ari_100_150:.3f} over {n_100_150} common hitters")
```

```python
from sklearn.cluster import KMeans

# =========================================
# 0) PARAMETERS
# =========================================
MIN_SWINGS = 100      # chosen based on my stability check (ARI ~ 0.98 for 100 vs 150)
N_CLUSTERS = 3
RANDOM_STATE = 42
N_INIT = 20

# =========================================
# 1) HITTER-LEVEL AGGREGATION
# =========================================
hitter_stats = (
    df_clean
    .group_by("batter")
    .agg([
        pl.count().alias("n_swings"),
        pl.mean("bat_speed").alias("avg_bat_speed"),
        pl.std("bat_speed").alias("std_bat_speed"),
        pl.mean("swing_length").alias("avg_swing_length"),
        pl.std("swing_length").alias("std_swing_length"),
        pl.mean("foul_occurred").alias("foul_rate_2strike"),
    ])
    .filter(pl.col("n_swings") >= MIN_SWINGS)  # justified by your ARI results
)

# (optional) peek at how many hitters youll cluster
print("Hitters eligible (>= {} swings):".format(MIN_SWINGS), hitter_stats.height)

```

```python
# =========================================
# 2) K-MEANS CLUSTERING (on means)
#    (optionally you could standardize; not necessary if youre comfortable with units)
# =========================================

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
features = ["avg_bat_speed","avg_swing_length"]

# enforce exposure floor before clustering (you said >=100)
hitter_stats = hitter_stats.filter(pl.col("n_swings") >= 100)

hit_pd = hitter_stats.select(["batter","n_swings"] + features).to_pandas()

scaler = StandardScaler()
Z = scaler.fit_transform(hit_pd[features].to_numpy())

kmeans = KMeans(n_clusters=N_CLUSTERS, n_init=N_INIT, random_state=RANDOM_STATE)
hit_pd["archetype"] = kmeans.fit_predict(Z)

# # --- centroids back to ORIGINAL units ---
# centroids = pd.DataFrame(
#     scaler.inverse_transform(kmeans.cluster_centers_),
#     columns=features
# )
# centroids["cluster"] = np.arange(len(centroids))

# print("\nCluster centroids (avg_bat_speed, avg_swing_length):\n", centroids)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import numpy as np
import pandas as pd

def fit_and_report(hit_pd, Z, scaler, k_list=(2,3,4,5,8), n_init=50, random_state=42,
                   features=("avg_bat_speed","avg_swing_length")):
    reports, details = [], {}
    n = len(Z)
    for K in k_list:
        if K < 2 or K >= n:  # safety
            continue
        km = KMeans(n_clusters=K, n_init=n_init, random_state=random_state)
        labels = km.fit_predict(Z)

        try:
            sil = silhouette_score(Z, labels)
            dbi = davies_bouldin_score(Z, labels)
            ch  = calinski_harabasz_score(Z, labels)
        except Exception:
            sil, dbi, ch = np.nan, np.nan, np.nan

        sz = pd.Series(labels).value_counts().sort_index()
        pct = (sz / sz.sum() * 100).round(1)

        centers = pd.DataFrame(
            scaler.inverse_transform(km.cluster_centers_), columns=list(features)
        )
        centers["cluster"] = range(K)

        reports.append({
            "K": K,
            "silhouette": sil,
            "davies_bouldin": dbi,
            "calinski_harabasz": ch,
            "inertia": km.inertia_,
            "min_cluster_pct": pct.min(),
            "median_cluster_pct": pct.median(),
            "max_cluster_pct": pct.max()
        })
        details[K] = {
            "sizes": pd.DataFrame({"cluster": range(K), "n": sz.values, "pct": pct.values}),
            "centroids": centers
        }

    summary = pd.DataFrame(reports).sort_values("K").reset_index(drop=True)
    print("\n== Summary across K ==")
    print(summary)
    for K in sorted(details):
        print(f"\n-- K={K} cluster sizes --")
        print(details[K]["sizes"])
        print(f"\n-- K={K} centroids (original units) --")
        print(details[K]["centroids"])
    return summary, details
```

```python
features = ["avg_bat_speed","avg_swing_length"]

# 1) Fit fine-grained clusters (K=8) on standardized space
K_FINE = 8
km8 = KMeans(n_clusters=K_FINE, n_init=N_INIT, random_state=RANDOM_STATE)
hit_pd["fine_cluster"] = km8.fit_predict(Z)
```

```python
# 2) Fine centroids in ORIGINAL units + sizes
centers8_orig = pd.DataFrame(
    scaler.inverse_transform(km8.cluster_centers_), columns=features
).assign(fine_cluster=np.arange(K_FINE))
fine_sizes = pd.Series(hit_pd["fine_cluster"]).value_counts().sort_index().rename("n")
centers8_orig = centers8_orig.merge(
    fine_sizes.reset_index().rename(columns={"index":"fine_cluster"}), on="fine_cluster"
)
```

```python
# 3) Merge fine clusters into 3 macros using Ward linkage on the K=8 centers

from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=3, linkage="ward")
macro_id = agg.fit_predict(km8.cluster_centers_)  # uses the standardized space implicitly
centers8_orig["macro_id"] = macro_id

# --- NEW: map fine -> macro ids ---
fine_to_macro = dict(zip(centers8_orig["fine_cluster"].astype(int),
                         centers8_orig["macro_id"].astype(int)))

# 4) Size-weighted macro centroids (in ORIGINAL units) for naming + sanity
macro_centers = (
    centers8_orig.groupby("macro_id")
    .apply(lambda g: pd.Series({
        "avg_bat_speed": np.average(g["avg_bat_speed"], weights=g["n"]),
        "avg_swing_length": np.average(g["avg_swing_length"], weights=g["n"]),
        "n_subclusters": len(g),
        "n_hitters": g["n"].sum()
    }))
    .reset_index()
)
```

```python
# ...after you created `macro_centers` via the groupby/np.average step:

# --- Unique naming using relative position in standardized space ---
macro_Z = scaler.transform(macro_centers[["avg_bat_speed","avg_swing_length"]])
macro_centers["z_bs"] = macro_Z[:, 0]
macro_centers["z_sl"] = macro_Z[:, 1]

macro_centers["score"] = macro_centers["z_bs"] + macro_centers["z_sl"]  # higher = more power-like
order = np.argsort(macro_centers["score"].values)  # 0=lowest(Contact), 2=highest(Power)

unique_map = {
    int(macro_centers.iloc[order[0]]["macro_id"]): "Contact",
    int(macro_centers.iloc[order[1]]["macro_id"]): "Balanced",
    int(macro_centers.iloc[order[2]]["macro_id"]): "Power",
}
macro_centers["macro_name"] = macro_centers["macro_id"].map(unique_map)

# map to hitters
id_to_name = dict(zip(macro_centers["macro_id"], macro_centers["macro_name"]))
hit_pd["macro_class"] = hit_pd["fine_cluster"].map(fine_to_macro).map(id_to_name)

# (then merge to df_clean, build sanity tables, etc.)
```

```python
# --- 1) Merge macro labels back to swing-level data ---
arch_pl = pl.from_pandas(
    hit_pd[["batter", "fine_cluster", "macro_class"]]
).with_columns(
    pl.col("fine_cluster").cast(pl.Int8),
    pl.col("macro_class").cast(pl.Categorical)
)
df_model = df_clean.join(arch_pl, on="batter", how="inner")
print("Pitch-level rows with macro labels:", df_model.height)
```

```python
# --- 2) Sanity: macro class sizes (by hitter) ---
macro_sizes = (
    pl.from_pandas(hit_pd)[["batter","macro_class"]]
    .unique()
    .group_by("macro_class")
    .agg(pl.count().alias("n_hitters"))
    .with_columns((pl.col("n_hitters")*100 / pl.col("n_hitters").sum())
                  .round(1).alias("pct_hitters"))
    .sort("n_hitters", descending=True)
)
print("\nMacro class sizes (hitters):\n", macro_sizes)

```

```python
# ===========================
# Fast VI GLMM for MLB swings
# (with archetypes + optional heterogeneity)
# ===========================
# import numpy as np
# import pandas as pd
# import matplotlib.pyplot as plt

# import pymc as pm
# import aesara
# import aesara.tensor as at

# from sklearn.preprocessing import OneHotEncoder, SplineTransformer

# ---------------------------------------------------------------------
# 0) Pull columns from df_model  pandas and enforce required columns
# ---------------------------------------------------------------------
pdf = df_model.select([
    "batter","pitcher","foul_occurred",
    "bat_speed","swing_length","macro_class",
    "release_speed","pfx_x","pfx_z","plate_x","plate_z",
    "inning","outs_when_up","balls","stand","p_throws",
    "game_date"
]).to_pandas()

# make sure required ones exist
req_cols = ["batter","pitcher","foul_occurred","bat_speed","swing_length","macro_class","game_date"]
missing = [c for c in req_cols if c not in pdf.columns]
assert len(missing)==0, f"Missing required columns: {missing}"

# ---------------------------------------------------------------------
# 1) One-hot encode archetypes (drop baseline automatically)
# ---------------------------------------------------------------------
try:
    ohe = OneHotEncoder(drop="first", sparse_output=False, dtype=np.float64)
except TypeError:
    ohe = OneHotEncoder(drop="first", sparse=False, dtype=np.float64)

A = ohe.fit_transform(pdf[["macro_class"]])   # (N, K-1)
arch_cols_raw = list(ohe.get_feature_names_out(["macro_class"]))   # e.g. ['macro_class_Contact','macro_class_Power']
arch_labels   = [c.split("_", 1)[1] for c in arch_cols_raw]        # ['Contact','Power']
A_cols        = [f"ARCH_{lab}" for lab in arch_labels]

# Baseline archetype (the one dropped by OHE):
all_arch = sorted(pd.Series(pdf["macro_class"]).dropna().unique().tolist())
baseline_arch = next((a for a in all_arch if a not in arch_labels), (all_arch[0] if all_arch else "Balanced"))

# ---------------------------------------------------------------------
# 2) Splines for bat_speed & swing_length (cubic, 6 knots)
# ---------------------------------------------------------------------
bs = SplineTransformer(degree=3, n_knots=6, include_bias=False)
sl = SplineTransformer(degree=3, n_knots=6, include_bias=False)

B = bs.fit_transform(pdf[["bat_speed"]].to_numpy())      # (N, pB)
L = sl.fit_transform(pdf[["swing_length"]].to_numpy())   # (N, pL)
B_cols = [f"BS_s{i}" for i in range(B.shape[1])]
L_cols = [f"SL_s{i}" for i in range(L.shape[1])]
```

```python
# ---------------------------------------------------------------------
# 3) Controls, context, handedness (numeric)
# ---------------------------------------------------------------------
controls = pdf[["release_speed","pfx_x","pfx_z","plate_x","plate_z"]].astype(np.float64).to_numpy()
ctrl_cols = ["release_speed","pfx_x","pfx_z","plate_x","plate_z"]

context = pdf[["inning","outs_when_up","balls"]].astype(np.int64).to_numpy()
ctx_cols = ["inning","outs_when_up","balls"]

stand_map, throw_map = {"R": 1, "L": -1}, {"R": 1, "L": -1}
stand_code = pdf["stand"].map(stand_map).astype(np.int8).to_numpy()[:, None]
throw_code = pdf["p_throws"].map(throw_map).astype(np.int8).to_numpy()[:, None]
hand_cols  = ["stand_code","throws_code"]
```

```python
# ---------------------------------------------------------------------
# 4) Row alignment mask (fixes your mismatch error)
#    Make ONE common mask across all blocks, slice everything once
# ---------------------------------------------------------------------
N0 = len(pdf)
arrays_for_mask = [
    B, L,
    controls.astype(np.float64, copy=False),
    context.astype(np.float64, copy=False),
    stand_code.astype(np.float64, copy=False),
    throw_code.astype(np.float64, copy=False),
]
common_mask = np.ones(N0, dtype=bool)
for arr in arrays_for_mask:
    if arr is not None and arr.size > 0:
        common_mask &= np.isfinite(arr).all(axis=1)

# also require required fields not missing
for c in ["foul_occurred","batter","pitcher","game_date","macro_class","bat_speed","swing_length"]:
    common_mask &= pdf[c].notna().to_numpy()

dropped = int((~common_mask).sum())
print(f"[Row alignment] Dropping rows due to any non-finite/missing across blocks: {dropped}")

pdf          = pdf.loc[common_mask].reset_index(drop=True)
A            = A[common_mask]
B            = B[common_mask]
L            = L[common_mask]
controls     = controls[common_mask]
context      = context[common_mask]
stand_code   = stand_code[common_mask]
throw_code   = throw_code[common_mask]

y                = pdf["foul_occurred"].astype(np.int8).to_numpy()
batter_raw       = pdf["batter"].astype("int64").to_numpy()
pitcher_raw      = pdf["pitcher"].astype("int64").to_numpy()
game_date_series = pd.to_datetime(pdf["game_date"], errors="coerce")
```

```python
# ---------------------------------------------------------------------
# 5) OPTIONAL heterogeneity: let curves differ by archetype
#    Set HETEROGENEITY=True to give each archetype its own BS/SL curve
# ---------------------------------------------------------------------
HETEROGENEITY = True

BSx_list, SLx_list = [], []
BSx_cols,  SLx_cols = [], []
if HETEROGENEITY and A.shape[1] > 0:
    for j, lab in enumerate(arch_labels):
        Aj = A[:, [j]]                  # (N,1) class dummy
        BSx_list.append(B * Aj)         # (N, pB)
        SLx_list.append(L * Aj)         # (N, pL)
        BSx_cols += [f"BS_s{i}:class_{lab}" for i in range(B.shape[1])]
        SLx_cols += [f"SL_s{i}:class_{lab}" for i in range(L.shape[1])]
```

```python

# ---------------------------------------------------------------------
# 6) Assemble X (exact order is mirrored in X_cols)
# ---------------------------------------------------------------------
X_blocks = [
    B,
    L,
    (np.hstack(BSx_list) if BSx_list else np.zeros((len(pdf), 0))),
    (np.hstack(SLx_list) if SLx_list else np.zeros((len(pdf), 0))),
    A.astype(np.float64),
    controls,
    context.astype(np.float64),
    stand_code.astype(np.float64),
    throw_code.astype(np.float64),
]
X = np.hstack(X_blocks).astype(np.float64)

X_cols = (
    B_cols
  + L_cols
  + BSx_cols
  + SLx_cols
  + A_cols
  + ctrl_cols
  + ctx_cols
  + hand_cols
)

print("Final aligned shapes:",
      f"B={B.shape}, L={L.shape}, A={A.shape},",
      f"controls={controls.shape}, context={context.shape},",
      f"stand={stand_code.shape}, throw={throw_code.shape}, X={X.shape}")
```

```python

# ---------------------------------------------------------------------
# 7) Standardize continuous columns; keep archetype dummies unscaled
# ---------------------------------------------------------------------
P = X.shape[1]
onehot_mask = np.array([name in A_cols for name in X_cols], dtype=bool)
cont_mask   = ~onehot_mask

X_mean = np.zeros(P, dtype=np.float64)
X_std  = np.ones(P,  dtype=np.float64)
if cont_mask.any():
    X_mean[cont_mask] = X[:, cont_mask].mean(axis=0)
    X_std[cont_mask]  = X[:, cont_mask].std(axis=0, ddof=0)
    X_std[cont_mask]  = np.where(X_std[cont_mask]==0.0, 1.0, X_std[cont_mask])

Xs = (X - X_mean) / X_std
Xs[:, onehot_mask] = X[:, onehot_mask]

# ---------------------------------------------------------------------
# 8) Densify IDs AFTER filtering
# ---------------------------------------------------------------------
def densify_ids(raw_ids: np.ndarray):
    levels, inv = np.unique(raw_ids, return_inverse=True)
    return inv.astype(np.int64, copy=False), levels

batter_index, batter_levels   = densify_ids(batter_raw)
pitcher_index, pitcher_levels = densify_ids(pitcher_raw)

# ---------------------------------------------------------------------
# 9) Chronological split by game week (future-style validation)
# ---------------------------------------------------------------------
rng = np.random.default_rng(123)
if game_date_series.notna().any():
    week_key   = game_date_series.dt.to_period("W").astype(str).to_numpy()
    uniq_weeks = np.array(sorted(pd.unique(week_key[pd.notna(week_key)])))
    if len(uniq_weeks) >= 5:
        cut = int(np.floor(0.8 * len(uniq_weeks)))  # 80/20 by weeks
        train_weeks = set(uniq_weeks[:cut]); test_weeks = set(uniq_weeks[cut:])
        is_test = np.array([w in test_weeks for w in week_key], dtype=bool)
    else:
        is_test = rng.random(X.shape[0]) < 0.2
else:
    is_test = rng.random(X.shape[0]) < 0.2
```

```python


# ---------------------------------------------------------------------
# 10) Sizes & prior center
# ---------------------------------------------------------------------
N, P = Xs.shape
y = np.asarray(y, dtype=np.int8)
n_bat = int(batter_index.max()) + 1
n_pit = int(pitcher_index.max()) + 1
print(f"N={N}, P={P}, n_bat={n_bat}, n_pit={n_pit}")

p_bar  = float(y.mean())
logit  = lambda p: np.log(p/(1-p))
mu_int = logit(np.clip(p_bar, 1e-6, 1-1e-6))

# ---------------------------------------------------------------------
# 11) Shared vars + minibatch
# ---------------------------------------------------------------------
Xs_sh = aesara.shared(Xs)
y_sh  = aesara.shared(y)
b_sh  = aesara.shared(batter_index)
p_sh  = aesara.shared(pitcher_index)
row_idx = pm.Minibatch(np.arange(N), batch_size=4096, random_seed=42)

# ---------------------------------------------------------------------
# 12) PyMC GLMM (logistic) with batter/pitcher random intercepts (ADVI)
#     (You can tighten priors later if you want more smoothing)
# ---------------------------------------------------------------------
with pm.Model() as model:
    intercept = pm.Normal("intercept", mu=mu_int, sigma=1.5)
    beta      = pm.Normal("beta", 0.0, 1.0, shape=P)

    sigma_bat = pm.HalfNormal("sigma_batter", sigma=1.0)
    sigma_pit = pm.HalfNormal("sigma_pitcher", sigma=1.0)
    z_bat     = pm.Normal("z_batter", 0.0, 1.0, shape=n_bat)
    z_pit     = pm.Normal("z_pitcher", 0.0, 1.0, shape=n_pit)
    bat_eff   = pm.Deterministic("batter_re", z_bat * sigma_bat)
    pit_eff   = pm.Deterministic("pitcher_re", z_pit * sigma_pit)

    i   = row_idx
    eta = intercept + at.dot(Xs_sh[i], beta) + bat_eff[b_sh[i]] + pit_eff[p_sh[i]]
    y_obs = pm.Bernoulli("y", logit_p=eta, observed=y_sh[i], total_size=N)

    approx   = pm.fit(n=30_000, method="advi")
    idata_vi = approx.sample(1000)

print("PyMC version:", pm.__version__)
```

```python


# ---------------------------------------------------------------------
# 13) Summaries
# ---------------------------------------------------------------------
post = idata_vi.posterior
intercept_mean = float(post["intercept"].mean(("chain","draw")))
beta_mean      = post["beta"].mean(("chain","draw")).values
sigma_bat_mean = float(post["sigma_batter"].mean(("chain","draw")))
sigma_pit_mean = float(post["sigma_pitcher"].mean(("chain","draw")))
print(f"Intercept mean: {intercept_mean:.3f}")
print(f"_batter: {sigma_bat_mean:.3f}   _pitcher: {sigma_pit_mean:.3f}")
```

```python
# ---------------------------------------------------------------------
# 14) Partial Dependence (curvy) by archetype  matches your original style
# ---------------------------------------------------------------------
def _knots(st):
    if hasattr(st, "knots_"): return np.unique(np.ravel(st.knots_)).astype(float)
    if hasattr(st, "bsplines_") and st.bsplines_: return np.unique(st.bsplines_[0].t).astype(float)
    return None
bs_knots = _knots(bs); sl_knots = _knots(sl)

# Keep archetype one-hots unscaled
onehot_mask = np.array([c in A_cols for c in X_cols], dtype=bool)
arch_to_idx = {lab: i for i, lab in enumerate(arch_labels)}

def _arch_vec(arche):
    v = np.zeros((1, len(arch_labels)), dtype=float)
    if arche in arch_to_idx:
        v[0, arch_to_idx[arche]] = 1.0
    return v

# Helper to know block lengths (for building PD rows in same order)
def slen(n): return n if isinstance(n, int) else n

pB, pL = B.shape[1], L.shape[1]
pA     = len(arch_labels)
pBSx   = pB * pA if (HETEROGENEITY and pA>0) else 0
pSLx   = pL * pA if (HETEROGENEITY and pA>0) else 0
pC     = controls.shape[1]
pK     = context.shape[1]
pH     = 2

# slices (not strictly needed, but helpful for audit)
# B | L | BSx | SLx | A | C | K | H
offsets = np.cumsum([0, pB, pL, pBSx, pSLx, pA, pC, pK, pH])
sB  = slice(offsets[0], offsets[0]+pB)
sL  = slice(offsets[1], offsets[1]+pL)
sBA = slice(offsets[2], offsets[2]+pBSx)
sLA = slice(offsets[3], offsets[3]+pSLx)
sA  = slice(offsets[4], offsets[4]+pA)
sC  = slice(offsets[5], offsets[5]+pC)
sK  = slice(offsets[6], offsets[6]+pK)
sH  = slice(offsets[7], offsets[7]+pH)

def _build_row(bs_val, sl_val, arche):
    """Create one design row in the SAME column order as X, then standardize it."""
    B_row = bs.transform([[bs_val]])        # (1, pB)
    L_row = sl.transform([[sl_val]])        # (1, pL)
    A_row = _arch_vec(arche)                # (1, pA)

    # interactions if enabled
    if HETEROGENEITY and pA>0:
        # expand BSarch
        if A_row.shape[1] > 0:
            BSx_parts = [B_row * A_row[0, j] for j in range(pA)]
            SLx_parts = [L_row * A_row[0, j] for j in range(pA)]
            BSx_row = np.hstack(BSx_parts)
            SLx_row = np.hstack(SLx_parts)
        else:
            BSx_row = np.zeros((1, 0)); SLx_row = np.zeros((1, 0))
    else:
        BSx_row = np.zeros((1, 0)); SLx_row = np.zeros((1, 0))

    # zeros for controls/context/handedness (population-average PD)
    C_row = np.zeros((1, pC))
    K_row = np.zeros((1, pK))
    H_row = np.zeros((1, pH))

    X_row = np.hstack([B_row, L_row, BSx_row, SLx_row, A_row, C_row, K_row, H_row])
    Xs_row = (X_row - X_mean) / X_std
    Xs_row[:, onehot_mask] = X_row[:, onehot_mask]   # keep ARCH as 0/1
    return Xs_row
```

### Counterfactual causal query (Controlled Direct Effect, + bat_speed)

Two flavors:

    Point CDE: hold all other features at medians/baseline; change only bat_speed from v0v1.
    
    Population CDE: average the effect over actual rows (covariates) for each archetype.

```python
# ===========================================
# Counterfactual CDEs (bat_speed, swing_length, plate_x, plate_z)
# Works whether plate_x/plate_z are spline'd or plain linear controls
# ===========================================
import numpy as np
import pandas as pd

def _sigmoid(z): 
    return 1.0/(1.0+np.exp(-z))

# ---------- 1) Recover block sizes from X_cols ----------
def _count(names, prefix, with_colon=False):
    if with_colon is None:          # exact match only
        return sum(n == prefix for n in names)
    if with_colon:                   # interaction columns contain ':'
        return sum(n.startswith(prefix) and (":" in n) for n in names)
    else:                            # base spline columns (no ':')
        return sum(n.startswith(prefix) and (":" not in n) for n in names)

pB  = _count(X_cols, "BS_s", with_colon=False)      # bat_speed bases
pL  = _count(X_cols, "SL_s", with_colon=False)      # swing_length bases
pBA = _count(X_cols, "BS_s", with_colon=True)       # BSARCH
pLA = _count(X_cols, "SL_s", with_colon=True)       # SLARCH
pA  = len(A_cols)                                   # ARCH mains

# remaining named blocks
def _has(n): return n in X_cols
C_names = [c for c in ["release_speed","pfx_x","pfx_z","plate_x","plate_z"] if _has(c)]
pC = len(C_names)
pK = sum(_has(c) for c in ["inning","outs_when_up","balls"])
pH = sum(_has(c) for c in ["stand_code","throws_code"])

# slices (must match your build order: B, L, (PX if used), (PZ if used), BA, LA, LI, A, C, K, H)
# Your grouped ridge build had no PX/PZ spline blocks and no LI block; we still keep placeholders.
sB  = slice(0, pB)
sL  = slice(sB.stop,  sB.stop + pL)
sPX = slice(sL.stop,  sL.stop + 0)        # 0 if you didn't spline plate_x
sPZ = slice(sPX.stop, sPX.stop + 0)       # 0 if you didn't spline plate_z
sBA = slice(sPZ.stop, sPZ.stop + pBA)
sLA = slice(sBA.stop, sBA.stop + pLA)
sLI = slice(sLA.stop, sLA.stop + 0)       # 0 if you didn't build a low-rank PXPZ
sA  = slice(sLI.stop, sLI.stop + pA)
sC  = slice(sA.stop,  sA.stop  + pC)
sK  = slice(sC.stop,  sC.stop  + pK)
sH  = slice(sK.stop,  sK.stop  + pH)

def slen(slc): 
    return 0 if (slc is None) else (slc.stop - slc.start)

# sanity check
P_expect = sum(slen(s) for s in [sB,sL,sPX,sPZ,sBA,sLA,sLI,sA,sC,sK,sH])
assert P_expect == len(X_cols), f"Block sizes ({P_expect}) != X_cols ({len(X_cols)})."

# ---------- 2) Archetype helpers ----------
arch_seen   = sorted(pd.Series(pdf["macro_class"]).dropna().astype(str).unique().tolist())
arch_nonbase = [c.replace("ARCH_","") for c in A_cols]
baseline_arch = next((a for a in arch_seen if f"ARCH_{a}" not in A_cols), arch_seen[0] if arch_seen else "Balanced")
arch_all_for_reports = [baseline_arch] + arch_nonbase
arch_to_idx = {lab:i for i,lab in enumerate(arch_nonbase)}

# keep archetype one-hots unscaled
onehot_mask = np.array([c in A_cols for c in X_cols], dtype=bool)

def canon_arch(name: str):
    """Normalize to the exact label in your data (handles 'Contact' vs 'class_Contact')."""
    if name is None: 
        return None
    name = str(name)
    cands = [name, ("class_"+name if not name.startswith("class_") else name.replace("class_","",1))]
    for c in cands:
        if c in arch_seen:
            return c
    return name

def _arch_vec(arche):
    v = np.zeros((1, len(arch_nonbase)), dtype=float)
    if arche in arch_to_idx:
        v[0, arch_to_idx[arche]] = 1.0
    return v

# For linear controls inside the C block, record their positions
C_pos = {name:i for i,name in enumerate(C_names)}

# ---------- 3) Build ONE standardized row in your exact block order ----------
# Notes:
#  - If you didn't spline plate_x/plate_z, keep PX/PZ blocks zero
#  - Use px_lin/pz_lin to set linear plate_x/plate_z inside the C block
def row_for_values(bs_val=None, sl_val=None, px_val=None, pz_val=None, px_lin=None, pz_lin=None, arche=None):
    B_row  = (bs.transform([[bs_val]]) if slen(sB)  else np.zeros((1,0))) if bs_val is not None else np.zeros((1, slen(sB)))
    L_row  = (sl.transform([[sl_val]]) if slen(sL)  else np.zeros((1,0))) if sl_val is not None else np.zeros((1, slen(sL)))
    PX_row = np.zeros((1, slen(sPX)))  # not using PX splines in your current build
    PZ_row = np.zeros((1, slen(sPZ)))  # not using PZ splines in your current build

    A_row  = _arch_vec(canon_arch(arche)) if slen(sA) else np.zeros((1,0))

    def _mult(base_row, out_len):
        if out_len == 0 or A_row.shape[1] == 0: 
            return np.zeros((1,0))
        return np.hstack([base_row * A_row[0,j] for j in range(A_row.shape[1])])

    BA_row = _mult(B_row, slen(sBA))
    LA_row = _mult(L_row, slen(sLA))
    LI_row = np.zeros((1, slen(sLI)))

    # C/K/H at standardized baseline = zeros  except let us set linear controls if requested
    C_row  = np.zeros((1, slen(sC)))
    if px_lin is not None and "plate_x" in C_pos:
        C_row[0, C_pos["plate_x"]] = float(px_lin)
    if pz_lin is not None and "plate_z" in C_pos:
        C_row[0, C_pos["plate_z"]] = float(pz_lin)

    K_row  = np.zeros((1, slen(sK)))
    H_row  = np.zeros((1, slen(sH)))

    X_row = np.hstack([B_row, L_row, PX_row, PZ_row, BA_row, LA_row, LI_row, A_row, C_row, K_row, H_row])
    Xs_row = (X_row - X_mean) / X_std
    Xs_row[:, onehot_mask] = X_row[:, onehot_mask]   # keep ARCH one-hots as 0/1
    return Xs_row  # shape (1,P)

# ---------- 4) Posterior draws (subset for speed) ----------
beta_draws = idata_vi.posterior["beta"].values.reshape(-1, len(X_cols))
int_draws  = idata_vi.posterior["intercept"].values.reshape(-1)
S = min(400, beta_draws.shape[0])
sel = np.linspace(0, beta_draws.shape[0]-1, S, dtype=int)
beta_samps = beta_draws[sel]
int_samps  = int_draws[sel]

# ---------- 5) Typical values (global medians or per-archetype) ----------
USE_ARCH_MEDIANS = False  # set True to hold class-specific medians if you prefer

def typical_vals(arche=None):
    if USE_ARCH_MEDIANS and arche is not None:
        a = canon_arch(arche)
        sub = pdf.loc[pdf["macro_class"].astype(str) == a]
        if len(sub) >= 10:
            return (float(sub["bat_speed"].median()),
                    float(sub["swing_length"].median()),
                    float(sub.get("plate_x", pd.Series([np.nan])).median()),
                    float(sub.get("plate_z", pd.Series([np.nan])).median()))
    # global medians
    return (float(pdf["bat_speed"].median()),
            float(pdf["swing_length"].median()),
            float(pdf.get("plate_x", pd.Series([np.nan])).median()),
            float(pdf.get("plate_z", pd.Series([np.nan])).median()))
# --- WHICH labels exist in the data, and which are used in the ARCH block ---
arch_seen = sorted(pd.Series(pdf["macro_class"]).dropna().astype(str).unique().tolist())
arch_nonbase = [c.replace("ARCH_","") for c in A_cols]   # e.g. ['Contact','Power']
arch_to_idx  = {lab:i for i,lab in enumerate(arch_nonbase)}
baseline_arch = next((a for a in arch_seen if f"ARCH_{a.replace('class_','')}" not in A_cols),
                     arch_seen[0] if arch_seen else "Balanced")

# Map any of {'Contact','class_Contact'} -> 'Contact' (the key used in A_cols)
def to_Ablock_label(name: str) -> str:
    s = str(name)
    return s.replace("class_","")

# Map any of {'Contact','class_Contact'} to the *exact label in the data* for filtering medians
# Preference: if the 'class_' version exists in data, use that; else use the raw.
def to_data_label(name: str) -> str:
    s = str(name)
    cand1 = s if s in arch_seen else None
    cand2 = ("class_"+s) if not s.startswith("class_") else s.replace("class_","",1)
    for c in (cand1, cand2):
        if c is not None and c in arch_seen:
            return c
    return s

# ARCH one-hots (non-baseline only)
def _arch_vec(arche) -> np.ndarray:
    v = np.zeros((1, len(arch_nonbase)), float)
    key = to_Ablock_label(arche)
    if key in arch_to_idx:
        v[0, arch_to_idx[key]] = 1.0
    return v

# Use per-class medians when USE_ARCH_MEDIANS=True
USE_ARCH_MEDIANS = True

def typical_vals(arche=None):
    # when arche is provided, compute medians for that class (if enough rows)
    if USE_ARCH_MEDIANS and arche is not None:
        a_data = to_data_label(arche)
        sub = pdf.loc[pdf["macro_class"].astype(str) == a_data]
        if len(sub) >= 10:   # require a little support
            bs_med = float(sub["bat_speed"].median())
            sl_med = float(sub["swing_length"].median())
            px_med = float(sub["plate_x"].median()) if "plate_x" in sub else np.nan
            pz_med = float(sub["plate_z"].median()) if "plate_z" in sub else np.nan
            return bs_med, sl_med, px_med, pz_med
    # otherwise global medians
    bs_med = float(pdf["bat_speed"].median())
    sl_med = float(pdf["swing_length"].median())
    px_med = float(pdf["plate_x"].median()) if "plate_x" in pdf else np.nan
    pz_med = float(pdf["plate_z"].median()) if "plate_z" in pdf else np.nan
    return bs_med, sl_med, px_med, pz_med

# ---------- 6) Generic CDE engines ----------
def point_cde(var, v0, v1, arche_list=None):
    """CDE at fixed typical values (medians), varying only var from v0->v1."""
    if arche_list is None:
        arche_list = arch_all_for_reports[:]
    out, diffs_all = {}, []

    for a in arche_list:
        bs_med, sl_med, px_med, pz_med = typical_vals(a)
        kw0 = dict(bs_val=bs_med, sl_val=sl_med, arche=a)
        kw1 = kw0.copy()
        if   var == "bat_speed":     kw0["bs_val"], kw1["bs_val"] = v0, v1
        elif var == "swing_length":  kw0["sl_val"], kw1["sl_val"] = v0, v1
        elif var == "plate_x":       kw0["px_lin"], kw1["px_lin"] = v0, v1   # linear control
        elif var == "plate_z":       kw0["pz_lin"], kw1["pz_lin"] = v0, v1   # linear control
        else: raise ValueError("var must be one of: bat_speed, swing_length, plate_x, plate_z")

        X0 = row_for_values(**kw0)
        X1 = row_for_values(**kw1)
        eta0 = int_samps + (beta_samps @ X0.T).ravel()
        eta1 = int_samps + (beta_samps @ X1.T).ravel()
        d = _sigmoid(eta1) - _sigmoid(eta0)  # (S,)
        diffs_all.append(d)
        out[canon_arch(a)] = dict(mean=float(d.mean()),
                                  lo=float(np.quantile(d,0.025)),
                                  hi=float(np.quantile(d,0.975)))

    # simple equal-weight ATE across listed archetypes
    D = np.vstack(diffs_all).mean(axis=0)
    ate = dict(mean=float(D.mean()),
               lo=float(np.quantile(D,0.025)),
               hi=float(np.quantile(D,0.975)))
    return ate, out


def population_cde(var, v0, v1, arche_list=None, n_rows=4000, seed=7):
    """Average CDE over observed covariates; only change the target var."""
    if arche_list is None:
        arche_list = arch_all_for_reports[:]
    rng = np.random.default_rng(seed)
    idx = rng.choice(len(pdf), size=min(n_rows, len(pdf)), replace=False)
    arch_series = pdf.iloc[idx]["macro_class"].astype(str).values
    bs_vals = pdf.iloc[idx]["bat_speed"].values
    sl_vals = pdf.iloc[idx]["swing_length"].values
    px_vals = pdf.iloc[idx]["plate_x"].values if "plate_x" in pdf else np.full(len(idx), np.nan)
    pz_vals = pdf.iloc[idx]["plate_z"].values if "plate_z" in pdf else np.full(len(idx), np.nan)

    out, diffs_all = {}, []
    for a in arche_list:
        a_c = canon_arch(a)
        mask = (arch_series == a_c)
        if not mask.any():
            out[a_c] = None
            continue

        rows0, rows1 = [], []
        for b, s, pxv, pzv in zip(bs_vals[mask], sl_vals[mask], px_vals[mask], pz_vals[mask]):
            kw0 = dict(bs_val=b, sl_val=s, arche=a_c)
            kw1 = dict(bs_val=b, sl_val=s, arche=a_c)
            if   var == "bat_speed":     kw0["bs_val"], kw1["bs_val"] = v0, v1
            elif var == "swing_length":  kw0["sl_val"], kw1["sl_val"] = v0, v1
            elif var == "plate_x":       kw0["px_lin"], kw1["px_lin"] = v0, v1
            elif var == "plate_z":       kw0["pz_lin"], kw1["pz_lin"] = v0, v1
            rows0.append(row_for_values(**kw0))
            rows1.append(row_for_values(**kw1))

        X0 = np.vstack(rows0);  X1 = np.vstack(rows1)
        eta0 = int_samps[:,None] + beta_samps @ X0.T
        eta1 = int_samps[:,None] + beta_samps @ X1.T
        d = _sigmoid(eta1) - _sigmoid(eta0)    # (S, n_mask)
        diffs_all.append(d)
        out[a_c] = dict(mean=float(d.mean()),
                        lo=float(np.quantile(d,0.025)),
                        hi=float(np.quantile(d,0.975)),
                        n_rows=int(mask.sum()))

    D = np.hstack(diffs_all) if diffs_all else np.zeros((len(int_samps),0))
    ate = dict(mean=float(D.mean()) if D.size else np.nan,
               lo=float(np.quantile(D,0.025)) if D.size else np.nan,
               hi=float(np.quantile(D,0.975)) if D.size else np.nan,
               n_rows=int(D.shape[1]))
    return ate, out

# ---------- 7) Quick run: Q25  Q75 for each variable ----------
vars_to_run = []
if "bat_speed"    in pdf: vars_to_run.append("bat_speed")
if "swing_length" in pdf: vars_to_run.append("swing_length")
if "plate_x"      in pdf: vars_to_run.append("plate_x")
if "plate_z"      in pdf: vars_to_run.append("plate_z")

for var in vars_to_run:
    q25, q75 = np.quantile(pdf[var], [0.25, 0.75])
    print(f"\n=== {var}: Q25Q75 ({q25:.2f}  {q75:.2f}) ===")
    ate_pt, by_arch_pt = point_cde(var, q25, q75)
    print(f"POINT CDE  ATE: {100*ate_pt['mean']:.2f} pp  (95% CI {100*ate_pt['lo']:.2f}, {100*ate_pt['hi']:.2f})")
    for a in arch_all_for_reports:
        v = by_arch_pt.get(canon_arch(a))
        if v:
            print(f"  {a:>12}: {100*v['mean']:.2f} pp  (95% CI {100*v['lo']:.2f}, {100*v['hi']:.2f})")

    ate_pop, by_arch_pop = population_cde(var, q25, q75)
    print(f"POP  CDE  ATE: {100*ate_pop['mean']:.2f} pp  (95% CI {100*ate_pop['lo']:.2f}, {100*ate_pop['hi']:.2f})")
    for a in arch_all_for_reports:
        v = by_arch_pop.get(canon_arch(a))
        if v:
            print(f"  {a:>12}: {100*v['mean']:.2f} pp  (95% CI {100*v['lo']:.2f}, {100*v['hi']:.2f})  [n={v['n_rows']}]")
```

```python
# ===========================================
# Forest plots of CDEs (point & population)
# ===========================================
import numpy as np
import matplotlib.pyplot as plt

def _summ_to_rows(var, q25, q75, which="point"):
    """Call CDE fn and return tidy rows: one per archetype + an 'ATE' row."""
    if which == "point":
        ate, by_arch = point_cde(var, q25, q75)
    elif which == "population":
        ate, by_arch = population_cde(var, q25, q75)
    else:
        raise ValueError("which must be 'point' or 'population'")

    rows = []
    # ATE first
    rows.append(dict(var=var, label="ATE", mean=100*ate["mean"], lo=100*ate["lo"], hi=100*ate["hi"]))
    # Then each archetype (only those present)
    for a in arch_all_for_reports:
        a2 = canon_arch(a)
        v = by_arch.get(a2)
        if v is None: 
            continue
        rows.append(dict(var=var, label=a2, mean=100*v["mean"], lo=100*v["lo"], hi=100*v["hi"]))
    return rows

def forest_plot(rows, title):
    """Simple forest plot from list[dict(mean, lo, hi, label)] (all in percentage points)."""
    labels = [r["label"] for r in rows]
    means  = np.array([r["mean"] for r in rows], float)
    los    = np.array([r["lo"]   for r in rows], float)
    his    = np.array([r["hi"]   for r in rows], float)
    y = np.arange(len(labels))

    plt.figure(figsize=(6, 0.5*len(labels) + 1.5))
    plt.errorbar(x=means, y=y, xerr=[means-los, his-means], fmt='o', capsize=3)
    plt.axvline(0.0, ls='--', alpha=0.4)
    plt.yticks(y, labels)
    plt.xlabel(" foul probability (percentage points)")
    plt.title(title)
    plt.tight_layout()
    plt.show()

def run_forest_for(var):
    q25, q75 = np.quantile(pdf[var], [0.25, 0.75])
    # POINT (medians held)
    rows_pt  = _summ_to_rows(var, q25, q75, which="point")
    forest_plot(rows_pt, f"{var}: CDE (Point)  Q25Q75  ({q25:.2f}{q75:.2f})")
    # POPULATION (averaged covariates)
    rows_pop = _summ_to_rows(var, q25, q75, which="population")
    forest_plot(rows_pop, f"{var}: CDE (Population)  Q25Q75  ({q25:.2f}{q75:.2f})")

# ---- Run for all four variables you just summarized ----
for var in ["bat_speed","swing_length","plate_x","plate_z"]:
    if var in pdf.columns:
        run_forest_for(var)
```

```python
# ===========================================
# Counterfactual CDEs (bat_speed, swing_length, plate_x, plate_z)
# Works whether plate_x/plate_z are spline'd or plain linear controls
# ===========================================
import numpy as np
import pandas as pd

def _sigmoid(z): 
    return 1.0/(1.0+np.exp(-z))

# ---------- 1) Recover block sizes from X_cols ----------
def _count(names, prefix, with_colon=False):
    if with_colon is None:          # exact match only
        return sum(n == prefix for n in names)
    if with_colon:                   # interaction columns contain ':'
        return sum(n.startswith(prefix) and (":" in n) for n in names)
    else:                            # base spline columns (no ':')
        return sum(n.startswith(prefix) and (":" not in n) for n in names)

pB  = _count(X_cols, "BS_s", with_colon=False)      # bat_speed bases
pL  = _count(X_cols, "SL_s", with_colon=False)      # swing_length bases
pBA = _count(X_cols, "BS_s", with_colon=True)       # BSARCH
pLA = _count(X_cols, "SL_s", with_colon=True)       # SLARCH
pA  = len(A_cols)                                   # ARCH mains

# remaining named blocks
def _has(n): return n in X_cols
C_names = [c for c in ["release_speed","pfx_x","pfx_z","plate_x","plate_z"] if _has(c)]
pC = len(C_names)
pK = sum(_has(c) for c in ["inning","outs_when_up","balls"])
pH = sum(_has(c) for c in ["stand_code","throws_code"])

# slices (must match your build order: B, L, (PX if used), (PZ if used), BA, LA, LI, A, C, K, H)
# Your grouped ridge build had no PX/PZ spline blocks and no LI block; we still keep placeholders.
sB  = slice(0, pB)
sL  = slice(sB.stop,  sB.stop + pL)
sPX = slice(sL.stop,  sL.stop + 0)        # 0 if you didn't spline plate_x
sPZ = slice(sPX.stop, sPX.stop + 0)       # 0 if you didn't spline plate_z
sBA = slice(sPZ.stop, sPZ.stop + pBA)
sLA = slice(sBA.stop, sBA.stop + pLA)
sLI = slice(sLA.stop, sLA.stop + 0)       # 0 if you didn't build a low-rank PXPZ
sA  = slice(sLI.stop, sLI.stop + pA)
sC  = slice(sA.stop,  sA.stop  + pC)
sK  = slice(sC.stop,  sC.stop  + pK)
sH  = slice(sK.stop,  sK.stop  + pH)

def slen(slc): 
    return 0 if (slc is None) else (slc.stop - slc.start)

# sanity check
P_expect = sum(slen(s) for s in [sB,sL,sPX,sPZ,sBA,sLA,sLI,sA,sC,sK,sH])
assert P_expect == len(X_cols), f"Block sizes ({P_expect}) != X_cols ({len(X_cols)})."

# ---------- 2) Archetype helpers ----------
arch_seen   = sorted(pd.Series(pdf["macro_class"]).dropna().astype(str).unique().tolist())
arch_nonbase = [c.replace("ARCH_","") for c in A_cols]
baseline_arch = next((a for a in arch_seen if f"ARCH_{a}" not in A_cols), arch_seen[0] if arch_seen else "Balanced")
arch_all_for_reports = [baseline_arch] + arch_nonbase
arch_to_idx = {lab:i for i,lab in enumerate(arch_nonbase)}

# keep archetype one-hots unscaled
onehot_mask = np.array([c in A_cols for c in X_cols], dtype=bool)

def canon_arch(name: str):
    """Normalize to the exact label in your data (handles 'Contact' vs 'class_Contact')."""
    if name is None: 
        return None
    name = str(name)
    cands = [name, ("class_"+name if not name.startswith("class_") else name.replace("class_","",1))]
    for c in cands:
        if c in arch_seen:
            return c
    return name

def _arch_vec(arche):
    v = np.zeros((1, len(arch_nonbase)), dtype=float)
    if arche in arch_to_idx:
        v[0, arch_to_idx[arche]] = 1.0
    return v

# For linear controls inside the C block, record their positions
C_pos = {name:i for i,name in enumerate(C_names)}

# ---------- 3) Build ONE standardized row in your exact block order ----------
# Notes:
#  - If you didn't spline plate_x/plate_z, keep PX/PZ blocks zero
#  - Use px_lin/pz_lin to set linear plate_x/plate_z inside the C block
def row_for_values(bs_val=None, sl_val=None, px_val=None, pz_val=None, px_lin=None, pz_lin=None, arche=None):
    B_row  = (bs.transform([[bs_val]]) if slen(sB)  else np.zeros((1,0))) if bs_val is not None else np.zeros((1, slen(sB)))
    L_row  = (sl.transform([[sl_val]]) if slen(sL)  else np.zeros((1,0))) if sl_val is not None else np.zeros((1, slen(sL)))
    PX_row = np.zeros((1, slen(sPX)))  # not using PX splines in your current build
    PZ_row = np.zeros((1, slen(sPZ)))  # not using PZ splines in your current build

    A_row  = _arch_vec(canon_arch(arche)) if slen(sA) else np.zeros((1,0))

    def _mult(base_row, out_len):
        if out_len == 0 or A_row.shape[1] == 0: 
            return np.zeros((1,0))
        return np.hstack([base_row * A_row[0,j] for j in range(A_row.shape[1])])

    BA_row = _mult(B_row, slen(sBA))
    LA_row = _mult(L_row, slen(sLA))
    LI_row = np.zeros((1, slen(sLI)))

    # C/K/H at standardized baseline = zeros  except let us set linear controls if requested
    C_row  = np.zeros((1, slen(sC)))
    if px_lin is not None and "plate_x" in C_pos:
        C_row[0, C_pos["plate_x"]] = float(px_lin)
    if pz_lin is not None and "plate_z" in C_pos:
        C_row[0, C_pos["plate_z"]] = float(pz_lin)

    K_row  = np.zeros((1, slen(sK)))
    H_row  = np.zeros((1, slen(sH)))

    X_row = np.hstack([B_row, L_row, PX_row, PZ_row, BA_row, LA_row, LI_row, A_row, C_row, K_row, H_row])
    Xs_row = (X_row - X_mean) / X_std
    Xs_row[:, onehot_mask] = X_row[:, onehot_mask]   # keep ARCH one-hots as 0/1
    return Xs_row  # shape (1,P)

# ---------- 4) Posterior draws (subset for speed) ----------
beta_draws = idata_vi.posterior["beta"].values.reshape(-1, len(X_cols))
int_draws  = idata_vi.posterior["intercept"].values.reshape(-1)
S = min(400, beta_draws.shape[0])
sel = np.linspace(0, beta_draws.shape[0]-1, S, dtype=int)
beta_samps = beta_draws[sel]
int_samps  = int_draws[sel]

# ---------- 5) Typical values (global medians or per-archetype) ----------
USE_ARCH_MEDIANS = False  # set True to hold class-specific medians if you prefer

def typical_vals(arche=None):
    if USE_ARCH_MEDIANS and arche is not None:
        a = canon_arch(arche)
        sub = pdf.loc[pdf["macro_class"].astype(str) == a]
        if len(sub) >= 10:
            return (float(sub["bat_speed"].median()),
                    float(sub["swing_length"].median()),
                    float(sub.get("plate_x", pd.Series([np.nan])).median()),
                    float(sub.get("plate_z", pd.Series([np.nan])).median()))
    # global medians
    return (float(pdf["bat_speed"].median()),
            float(pdf["swing_length"].median()),
            float(pdf.get("plate_x", pd.Series([np.nan])).median()),
            float(pdf.get("plate_z", pd.Series([np.nan])).median()))
# --- WHICH labels exist in the data, and which are used in the ARCH block ---
arch_seen = sorted(pd.Series(pdf["macro_class"]).dropna().astype(str).unique().tolist())
arch_nonbase = [c.replace("ARCH_","") for c in A_cols]   # e.g. ['Contact','Power']
arch_to_idx  = {lab:i for i,lab in enumerate(arch_nonbase)}
baseline_arch = next((a for a in arch_seen if f"ARCH_{a.replace('class_','')}" not in A_cols),
                     arch_seen[0] if arch_seen else "Balanced")

# Map any of {'Contact','class_Contact'} -> 'Contact' (the key used in A_cols)
def to_Ablock_label(name: str) -> str:
    s = str(name)
    return s.replace("class_","")

# Map any of {'Contact','class_Contact'} to the *exact label in the data* for filtering medians
# Preference: if the 'class_' version exists in data, use that; else use the raw.
def to_data_label(name: str) -> str:
    s = str(name)
    cand1 = s if s in arch_seen else None
    cand2 = ("class_"+s) if not s.startswith("class_") else s.replace("class_","",1)
    for c in (cand1, cand2):
        if c is not None and c in arch_seen:
            return c
    return s

# ARCH one-hots (non-baseline only)
def _arch_vec(arche) -> np.ndarray:
    v = np.zeros((1, len(arch_nonbase)), float)
    key = to_Ablock_label(arche)
    if key in arch_to_idx:
        v[0, arch_to_idx[key]] = 1.0
    return v

# Use per-class medians when USE_ARCH_MEDIANS=True
USE_ARCH_MEDIANS = True

def typical_vals(arche=None):
    # when arche is provided, compute medians for that class (if enough rows)
    if USE_ARCH_MEDIANS and arche is not None:
        a_data = to_data_label(arche)
        sub = pdf.loc[pdf["macro_class"].astype(str) == a_data]
        if len(sub) >= 10:   # require a little support
            bs_med = float(sub["bat_speed"].median())
            sl_med = float(sub["swing_length"].median())
            px_med = float(sub["plate_x"].median()) if "plate_x" in sub else np.nan
            pz_med = float(sub["plate_z"].median()) if "plate_z" in sub else np.nan
            return bs_med, sl_med, px_med, pz_med
    # otherwise global medians
    bs_med = float(pdf["bat_speed"].median())
    sl_med = float(pdf["swing_length"].median())
    px_med = float(pdf["plate_x"].median()) if "plate_x" in pdf else np.nan
    pz_med = float(pdf["plate_z"].median()) if "plate_z" in pdf else np.nan
    return bs_med, sl_med, px_med, pz_med

# ---------- 6) Generic CDE engines ----------
def point_cde(var, v0, v1, arche_list=None):
    """CDE at fixed typical values (medians), varying only var from v0->v1."""
    if arche_list is None:
        arche_list = arch_all_for_reports[:]
    out, diffs_all = {}, []

    for a in arche_list:
        bs_med, sl_med, px_med, pz_med = typical_vals(a)
        kw0 = dict(bs_val=bs_med, sl_val=sl_med, arche=a)
        kw1 = kw0.copy()
        if   var == "bat_speed":     kw0["bs_val"], kw1["bs_val"] = v0, v1
        elif var == "swing_length":  kw0["sl_val"], kw1["sl_val"] = v0, v1
        elif var == "plate_x":       kw0["px_lin"], kw1["px_lin"] = v0, v1   # linear control
        elif var == "plate_z":       kw0["pz_lin"], kw1["pz_lin"] = v0, v1   # linear control
        else: raise ValueError("var must be one of: bat_speed, swing_length, plate_x, plate_z")

        X0 = row_for_values(**kw0)
        X1 = row_for_values(**kw1)
        eta0 = int_samps + (beta_samps @ X0.T).ravel()
        eta1 = int_samps + (beta_samps @ X1.T).ravel()
        d = _sigmoid(eta1) - _sigmoid(eta0)  # (S,)
        diffs_all.append(d)
        out[canon_arch(a)] = dict(mean=float(d.mean()),
                                  lo=float(np.quantile(d,0.025)),
                                  hi=float(np.quantile(d,0.975)))

    # simple equal-weight ATE across listed archetypes
    D = np.vstack(diffs_all).mean(axis=0)
    ate = dict(mean=float(D.mean()),
               lo=float(np.quantile(D,0.025)),
               hi=float(np.quantile(D,0.975)))
    return ate, out


def population_cde(var, v0, v1, arche_list=None, n_rows=4000, seed=7):
    """Average CDE over observed covariates; only change the target var."""
    if arche_list is None:
        arche_list = arch_all_for_reports[:]
    rng = np.random.default_rng(seed)
    idx = rng.choice(len(pdf), size=min(n_rows, len(pdf)), replace=False)
    arch_series = pdf.iloc[idx]["macro_class"].astype(str).values
    bs_vals = pdf.iloc[idx]["bat_speed"].values
    sl_vals = pdf.iloc[idx]["swing_length"].values
    px_vals = pdf.iloc[idx]["plate_x"].values if "plate_x" in pdf else np.full(len(idx), np.nan)
    pz_vals = pdf.iloc[idx]["plate_z"].values if "plate_z" in pdf else np.full(len(idx), np.nan)

    out, diffs_all = {}, []
    for a in arche_list:
        a_c = canon_arch(a)
        mask = (arch_series == a_c)
        if not mask.any():
            out[a_c] = None
            continue

        rows0, rows1 = [], []
        for b, s, pxv, pzv in zip(bs_vals[mask], sl_vals[mask], px_vals[mask], pz_vals[mask]):
            kw0 = dict(bs_val=b, sl_val=s, arche=a_c)
            kw1 = dict(bs_val=b, sl_val=s, arche=a_c)
            if   var == "bat_speed":     kw0["bs_val"], kw1["bs_val"] = v0, v1
            elif var == "swing_length":  kw0["sl_val"], kw1["sl_val"] = v0, v1
            elif var == "plate_x":       kw0["px_lin"], kw1["px_lin"] = v0, v1
            elif var == "plate_z":       kw0["pz_lin"], kw1["pz_lin"] = v0, v1
            rows0.append(row_for_values(**kw0))
            rows1.append(row_for_values(**kw1))

        X0 = np.vstack(rows0);  X1 = np.vstack(rows1)
        eta0 = int_samps[:,None] + beta_samps @ X0.T
        eta1 = int_samps[:,None] + beta_samps @ X1.T
        d = _sigmoid(eta1) - _sigmoid(eta0)    # (S, n_mask)
        diffs_all.append(d)
        out[a_c] = dict(mean=float(d.mean()),
                        lo=float(np.quantile(d,0.025)),
                        hi=float(np.quantile(d,0.975)),
                        n_rows=int(mask.sum()))

    D = np.hstack(diffs_all) if diffs_all else np.zeros((len(int_samps),0))
    ate = dict(mean=float(D.mean()) if D.size else np.nan,
               lo=float(np.quantile(D,0.025)) if D.size else np.nan,
               hi=float(np.quantile(D,0.975)) if D.size else np.nan,
               n_rows=int(D.shape[1]))
    return ate, out

# ---------- 7) Quick run: Q25  Q75 for each variable ----------
vars_to_run = []
if "bat_speed"    in pdf: vars_to_run.append("bat_speed")
if "swing_length" in pdf: vars_to_run.append("swing_length")
if "plate_x"      in pdf: vars_to_run.append("plate_x")
if "plate_z"      in pdf: vars_to_run.append("plate_z")

for var in vars_to_run:
    q25, q75 = np.quantile(pdf[var], [0.25, 0.75])
    print(f"\n=== {var}: Q25Q75 ({q25:.2f}  {q75:.2f}) ===")
    ate_pt, by_arch_pt = point_cde(var, q25, q75)
    print(f"POINT CDE  ATE: {100*ate_pt['mean']:.2f} pp  (95% CI {100*ate_pt['lo']:.2f}, {100*ate_pt['hi']:.2f})")
    for a in arch_all_for_reports:
        v = by_arch_pt.get(canon_arch(a))
        if v:
            print(f"  {a:>12}: {100*v['mean']:.2f} pp  (95% CI {100*v['lo']:.2f}, {100*v['hi']:.2f})")

    ate_pop, by_arch_pop = population_cde(var, q25, q75)
    print(f"POP  CDE  ATE: {100*ate_pop['mean']:.2f} pp  (95% CI {100*ate_pop['lo']:.2f}, {100*ate_pop['hi']:.2f})")
    for a in arch_all_for_reports:
        v = by_arch_pop.get(canon_arch(a))
        if v:
            print(f"  {a:>12}: {100*v['mean']:.2f} pp  (95% CI {100*v['lo']:.2f}, {100*v['hi']:.2f})  [n={v['n_rows']}]")
```

```python


# ===========================================
# Forest plots of CDEs (point & population)
# ===========================================
import numpy as np
import matplotlib.pyplot as plt

def _summ_to_rows(var, q25, q75, which="point"):
    """Call CDE fn and return tidy rows: one per archetype + an 'ATE' row."""
    if which == "point":
        ate, by_arch = point_cde(var, q25, q75)
    elif which == "population":
        ate, by_arch = population_cde(var, q25, q75)
    else:
        raise ValueError("which must be 'point' or 'population'")

    rows = []
    # ATE first
    rows.append(dict(var=var, label="ATE", mean=100*ate["mean"], lo=100*ate["lo"], hi=100*ate["hi"]))
    # Then each archetype (only those present)
    for a in arch_all_for_reports:
        a2 = canon_arch(a)
        v = by_arch.get(a2)
        if v is None: 
            continue
        rows.append(dict(var=var, label=a2, mean=100*v["mean"], lo=100*v["lo"], hi=100*v["hi"]))
    return rows

def forest_plot(rows, title):
    """Simple forest plot from list[dict(mean, lo, hi, label)] (all in percentage points)."""
    labels = [r["label"] for r in rows]
    means  = np.array([r["mean"] for r in rows], float)
    los    = np.array([r["lo"]   for r in rows], float)
    his    = np.array([r["hi"]   for r in rows], float)
    y = np.arange(len(labels))

    plt.figure(figsize=(6, 0.5*len(labels) + 1.5))
    plt.errorbar(x=means, y=y, xerr=[means-los, his-means], fmt='o', capsize=3)
    plt.axvline(0.0, ls='--', alpha=0.4)
    plt.yticks(y, labels)
    plt.xlabel(" foul probability (percentage points)")
    plt.title(title)
    plt.tight_layout()
    plt.show()

def run_forest_for(var):
    q25, q75 = np.quantile(pdf[var], [0.25, 0.75])
    # POINT (medians held)
    rows_pt  = _summ_to_rows(var, q25, q75, which="point")
    forest_plot(rows_pt, f"{var}: CDE (Point)  Q25Q75  ({q25:.2f}{q75:.2f})")
    # POPULATION (averaged covariates)
    rows_pop = _summ_to_rows(var, q25, q75, which="population")
    forest_plot(rows_pop, f"{var}: CDE (Population)  Q25Q75  ({q25:.2f}{q75:.2f})")

# ---- Run for all four variables you just summarized ----
for var in ["bat_speed","swing_length","plate_x","plate_z"]:
    if var in pdf.columns:
        run_forest_for(var)
```

#### Uncertainty presentation
(i) PD curves with 95% credible bands, by archetype

This uses your spline transformers and re-creates rows in the exact block order you used. It automatically handles presence/absence of PX/PZ splines and one-hot archetypes.

```python
# ===== PD with uncertainty bands (bat_speed & swing_length) =====
def _has_block(name): return any(c.startswith(name) for c in X_cols)
def _count_plain(prefix):  return sum((c.startswith(prefix) and ":" not in c) for c in X_cols)
def _count_xarch(prefix):  return sum((c.startswith(prefix) and ":" in c)  for c in X_cols)

pB, pL = _count_plain("BS_s"), _count_plain("SL_s")
pBA, pLA = _count_xarch("BS_s"), _count_xarch("SL_s")
pA = len(A_cols)
pPX = _count_plain("PX_s"); pPZ = _count_plain("PZ_s")

# slices in your block order
i=0; sB=slice(i,i+pB); i=sB.stop
sL=slice(i,i+pL); i=sL.stop
sPX=slice(i,i+pPX); i=sPX.stop
sPZ=slice(i,i+pPZ); i=sPZ.stop
sBA=slice(i,i+pBA); i=sBA.stop
sLA=slice(i,i+pLA); i=sLA.stop
sA =slice(i,i+pA);  i=sA.stop
# controls/context/handedness the rest:
tail = len(X_cols) - i
# split tail into C,K,H by membership:
C_cols = [c for c in ["release_speed","pfx_x","pfx_z","plate_x","plate_z"] if c in X_cols]
K_cols = [c for c in ["inning","outs_when_up","balls"] if c in X_cols]
H_cols = [c for c in ["stand_code","throws_code"]     if c in X_cols]
sC = slice(i, i+len(C_cols)); i=sC.stop
sK = slice(i, i+len(K_cols)); i=sK.stop
sH = slice(i, i+len(H_cols)); i=sH.stop
assert i == len(X_cols), "Slices must cover all columns."

onehot_mask = np.array([c in A_cols for c in X_cols], bool)
arch_nonbase = [c.replace("ARCH_","") for c in A_cols]
arch_all = sorted(pd.Series(pdf["macro_class"]).dropna().astype(str).unique())
baseline_arch = next((a for a in arch_all if f"ARCH_{a}" not in A_cols), (arch_all[0] if arch_all else "Balanced"))
arch_display = [baseline_arch] + arch_nonbase
arch_to_idx = {lab:i for i,lab in enumerate(arch_nonbase)}

def _arch_row(a):
    v = np.zeros((1,len(arch_nonbase)))
    if a in arch_to_idx: v[0, arch_to_idx[a]] = 1.0
    return v

def _row(bs_val=None, sl_val=None, px_val=None, pz_val=None, arche=None):
    blocks = []
    # bases
    B = bs.transform([[bs_val]]) if pB>0 else np.zeros((1,0))
    L = sl.transform([[sl_val]]) if pL>0 else np.zeros((1,0))
    blocks += [B, L]
    # plate splines if you built them
    if pPX>0:
        assert 'px' in globals(), "px transformer not found"
        blocks.append(px.transform([[px_val]]))
    else:
        blocks.append(np.zeros((1,0)))
    if pPZ>0:
        assert 'pz' in globals(), "pz transformer not found"
        blocks.append(pz.transform([[pz_val]]))
    else:
        blocks.append(np.zeros((1,0)))
    # arche main + interactions
    A = _arch_row(arche) if pA>0 else np.zeros((1,0))
    def _xarch(base, pX):
        if (pX==0) or (A.shape[1]==0): return np.zeros((1,0))
        return np.hstack([base*A[0,j] for j in range(A.shape[1])])
    BA = _xarch(B, pBA); LA = _xarch(L, pLA)
    blocks += [BA, LA, A]
    # controls / context / handedness at standardized baseline (zeros post-zscore)
    blocks += [np.zeros((1,len(C_cols))), np.zeros((1,len(K_cols))), np.zeros((1,len(H_cols)))]
    X_row = np.hstack(blocks)
    Xs_row = (X_row - X_mean)/X_std
    Xs_row[:,onehot_mask] = X_row[:,onehot_mask]
    return Xs_row

# posterior subset
beta_draws = idata_vi.posterior["beta"].values.reshape(-1, len(X_cols))
int_draws  = idata_vi.posterior["intercept"].values.reshape(-1)
S = min(400, beta_draws.shape[0]); take = np.linspace(0, beta_draws.shape[0]-1, S, dtype=int)
BETA = beta_draws[take]; INT = int_draws[take]

def pd_band(var="bat_speed", n=120):
    assert var in ["bat_speed","swing_length"]
    # fix covariates at medians for PD
    sl_med = float(pdf["swing_length"].median())
    bs_med = float(pdf["bat_speed"].median())
    px_med = float(pdf["plate_x"].median()) if "plate_x" in pdf else None
    pz_med = float(pdf["plate_z"].median()) if "plate_z" in pdf else None

    if var=="bat_speed":
        lo,hi = np.quantile(pdf["bat_speed"], [0.02,0.98])
        grid = np.linspace(lo,hi,n)
        def _build(a, v): return _row(bs_val=v, sl_val=sl_med, px_val=px_med, pz_val=pz_med, arche=a)
    else:
        lo,hi = np.quantile(pdf["swing_length"], [0.02,0.98])
        grid = np.linspace(lo,hi,n)
        def _build(a, v): return _row(bs_val=bs_med, sl_val=v, px_val=px_med, pz_val=pz_med, arche=a)

    out = {}
    for a in arch_display:
        Xgrid = np.vstack([_build(a,v) for v in grid])
        # population-average (no REs in PD)
        ETA = INT[:,None] + BETA @ Xgrid.T          # (S,n)
        P   = _sigmoid(ETA)
        mean = P.mean(axis=0); lo95 = np.quantile(P,0.025,axis=0); hi95 = np.quantile(P,0.975,axis=0)
        out[a] = (grid, mean, lo95, hi95)

    # plot
    import matplotlib.pyplot as plt
    plt.figure(figsize=(7,4))
    for a,(g,m,l,h) in out.items():
        plt.plot(g, m, label=a)
        plt.fill_between(g, l, h, alpha=.15)
    plt.xlabel("Bat speed" if var=="bat_speed" else "Swing length")
    plt.ylabel("Predicted foul probability")
    plt.title(f"PD with 95% bands  {var}")
    plt.legend(title="Archetype", frameon=False); plt.tight_layout(); plt.show()
    return out

_ = pd_band("bat_speed")
_ = pd_band("swing_length")
```

ii) Forest plots: per-archetype population CDEs (Q25Q75) with 95% CIs

(Works for bat_speed, swing_length, and, if you splined them, plate_x, plate_z.)

```python
# ===== Forest plots of archetype CDEs (population version) =====
def _canon_arch_list():
    arch_all = sorted(pd.Series(pdf["macro_class"]).dropna().astype(str).unique())
    base = next((a for a in arch_all if f"ARCH_{a}" not in A_cols), (arch_all[0] if arch_all else "Balanced"))
    return [base] + [c.replace("ARCH_","") for c in A_cols]

def cde_population(var, q0, q1, n_rows=4000, seed=7):
    rng = np.random.default_rng(seed)
    idx = rng.choice(len(pdf), size=min(n_rows, len(pdf)), replace=False)
    arch_series = pdf.iloc[idx]["macro_class"].astype(str).values
    bs_vals = pdf.iloc[idx]["bat_speed"].values
    sl_vals = pdf.iloc[idx]["swing_length"].values
    px_vals = pdf.iloc[idx]["plate_x"].values if "plate_x" in pdf else np.full(len(idx), np.nan)
    pz_vals = pdf.iloc[idx]["plate_z"].values if "plate_z" in pdf else np.full(len(idx), np.nan)

    beta_draws = idata_vi.posterior["beta"].values.reshape(-1, len(X_cols))
    int_draws  = idata_vi.posterior["intercept"].values.reshape(-1)
    S = min(400, beta_draws.shape[0]); take = np.linspace(0, beta_draws.shape[0]-1, S, dtype=int)
    B = beta_draws[take]; I = int_draws[take]

    def _pred_row(a, b, s, pxv, pzv, vvar):
        kw = dict(bs_val=b, sl_val=s,
                  px_val=None if (np.isnan(pxv) if isinstance(pxv,float) else False) else pxv,
                  pz_val=None if (np.isnan(pzv) if isinstance(pzv,float) else False) else pzv,
                  arche=a)
        if   var=="bat_speed":    X0=_row(q0, s, pxv, pzv, a); X1=_row(q1, s, pxv, pzv, a)
        elif var=="swing_length": X0=_row(b, q0, pxv, pzv, a); X1=_row(b, q1, pxv, pzv, a)
        elif var=="plate_x":      X0=_row(b, s, q0, pzv, a);    X1=_row(b, s, q1, pzv, a)
        elif var=="plate_z":      X0=_row(b, s, pxv, q0, a);    X1=_row(b, s, pxv, q1, a)
        else: raise ValueError
        return X0, X1

    results = {}
    for a in _canon_arch_list():
        m = (arch_series == a)
        if not m.any():
            results[a] = dict(mean=np.nan, lo=np.nan, hi=np.nan, n=0); continue
        rows0, rows1 = [], []
        for b,s,pxv,pzv in zip(bs_vals[m], sl_vals[m], px_vals[m], pz_vals[m]):
            X0, X1 = _pred_row(a, b, s, pxv, pzv, var)
            rows0.append(X0); rows1.append(X1)
        X0 = np.vstack(rows0); X1 = np.vstack(rows1)          # (n,P)
        ETA0 = I[:,None] + B @ X0.T; ETA1 = I[:,None] + B @ X1.T
        D = 1/(1+np.exp(-ETA1)) - 1/(1+np.exp(-ETA0))         # (S,n)
        results[a] = dict(mean=float(D.mean()),
                          lo=float(np.quantile(D,0.025)),
                          hi=float(np.quantile(D,0.975)),
                          n=int(m.sum()))
    return results

# run and plot for all variables present
vars_to_run = ["bat_speed","swing_length"]
if any(c.startswith("PX_s") for c in X_cols): vars_to_run.append("plate_x")
if any(c.startswith("PZ_s") for c in X_cols): vars_to_run.append("plate_z")

for var in vars_to_run:
    q25, q75 = np.quantile(pdf[var], [0.25,0.75])
    res = cde_population(var, q25, q75)
    # forest plot
    labs = list(res.keys())
    means = np.array([100*res[k]["mean"] for k in labs])
    los   = np.array([100*res[k]["lo"] for k in labs])
    his   = np.array([100*res[k]["hi"] for k in labs])

    order = np.argsort(means)
    labs_o = [labs[i] for i in order]; means_o = means[order]; los_o = los[order]; his_o = his[order]

    plt.figure(figsize=(6, 0.6*len(labs_o)+1))
    y = np.arange(len(labs_o))
    plt.errorbar(means_o, y, xerr=[means_o-los_o, his_o-means_o], fmt='o', capsize=3)
    plt.axvline(0, ls='--', alpha=.5)
    plt.yticks(y, labs_o)
    plt.xlabel(" foul probability (pp) for Q25Q75 of " + var)
    plt.title(f"Population CDE by archetype  {var}")
    plt.tight_layout(); plt.show()
```

```python
# --- Robust, case-insensitive archetype canonicalization ---

# all labels as they appear in your dataframe
arch_seen = sorted(pd.Series(pdf["macro_class"]).dropna().astype(str).unique().tolist())

# build a "core" -> canonical label map (prefer class_* form if both exist)
core_to_label = {}
for lab in arch_seen:
    core = lab.lower().replace("class_", "")
    # prefer the class_* version when there is a clash
    if core not in core_to_label or lab.startswith("class_"):
        core_to_label[core] = lab

def canon_arch(name: str):
    """Return the canonical label as it appears in pdf['macro_class']."""
    if name is None:
        return None
    core = str(name).lower().replace("class_", "")
    return core_to_label.get(core, None)

# baseline = the one not in A_cols (after canon)
A_nonbase = [c.replace("ARCH_", "") for c in A_cols]
A_nonbase_core = {a.lower().replace("class_", "") for a in A_nonbase}
baseline_arch = None
for lab in arch_seen:
    if lab.lower().replace("class_", "") not in A_nonbase_core:
        baseline_arch = lab
        break
if baseline_arch is None:                       # fallback
    baseline_arch = arch_seen[0]

arch_all_for_reports = [baseline_arch] + [canon_arch(a) for a in A_nonbase]
arch_all_for_reports = [a for a in arch_all_for_reports if a is not None]

def normalize_by_arch_keys(d):
    """Return dict whose keys are the canonical labels; drop empties."""
    out = {}
    for k, v in d.items():
        ck = canon_arch(k)
        if ck is not None:
            out[ck] = v
    return out

def gather_by_arch(by_arch_dict):
    by_arch_dict = normalize_by_arch_keys(by_arch_dict)
    means, los, his, names, present = [], [], [], [], []
    for a in arch_all_for_reports:
        v = by_arch_dict.get(a, None)
        if v is None:
            means.append(np.nan); los.append(np.nan); his.append(np.nan); present.append(False)
        else:
            means.append(100.0 * float(v["mean"]))
            los.append(100.0 * float(v["lo"]))
            his.append(100.0 * float(v["hi"]))
            present.append(True)
        names.append(a)
    return np.array(means), np.array(los), np.array(his), names, np.array(present, bool)

def plot_forest(by_arch_dict, title, xlab=" foul probability (pp)"):
    means, los, his, names, present = gather_by_arch(by_arch_dict)
    y = np.arange(len(names))[::-1]

    plt.figure(figsize=(7, 4.6))
    plt.axvline(0, ls="--", color="gray", alpha=0.6)
    for i, ok in enumerate(present):
        if not ok: 
            continue
        plt.hlines(y[i], xmin=los[i], xmax=his[i], color="C0", lw=2)
        plt.plot(means[i], y[i], "o", color="C0")
    plt.yticks(y, names)
    plt.xlabel(xlab); plt.title(title)
    # auto x-lims
    finite = np.isfinite(los) & np.isfinite(his)
    if finite.any():
        lo_x = float(np.nanmin(los[finite])); hi_x = float(np.nanmax(his[finite]))
        pad = 0.1 * (hi_x - lo_x if hi_x > lo_x else 1.0)
        plt.xlim(lo_x - pad, hi_x + pad)
    plt.grid(axis="x", alpha=0.15); plt.tight_layout(); plt.show()

# ---------- RE-RUN the four variables (use all rows so each class is present) ----------
vars_to_run = []
if "bat_speed"    in pdf.columns: vars_to_run.append("bat_speed")
if "swing_length" in pdf.columns: vars_to_run.append("swing_length")
if "plate_x"      in pdf.columns: vars_to_run.append("plate_x")
if "plate_z"      in pdf.columns: vars_to_run.append("plate_z")

for var in vars_to_run:
    q25, q75 = np.quantile(pdf[var], [0.25, 0.75])

    # force all archetypes; use every row for population CDE so none drop out
    ate_pt,  by_arch_pt  = point_cde(var, q25, q75, arche_list=arch_all_for_reports)
    ate_pop, by_arch_pop = population_cde(var, q25, q75, arche_list=arch_all_for_reports, n_rows=len(pdf), seed=7)

    print(f"\n{var} POINT keys:", sorted(normalize_by_arch_keys(by_arch_pt).keys()))
    print(f"{var} POP   keys:", sorted(normalize_by_arch_keys(by_arch_pop).keys()))

    plot_forest(by_arch_pt,  f"Point CDE by archetype  {var} (Q25Q75)")
    plot_forest(by_arch_pop, f"Population CDE by archetype  {var} (Q25Q75)")
```
